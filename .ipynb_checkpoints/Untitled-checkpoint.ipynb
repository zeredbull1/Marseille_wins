```python
import pandas as pd
import numpy as np

fields = ["Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR"]

df = pd.read_csv("data/Ligue1_game_stats.csv", usecols = fields)

df.head()
```

```python
# Dataset with only Marseille games

df_om_games = df[(df["HomeTeam"] == "Marseille") | (df["AwayTeam"] == "Marseille")]

# Marseille goals column
# FTHG and HG = Full Time Home Team Goals
# FTAG and AG = Full Time Away Team Goals
# FTR and Res = Full Time Result (H=Home Win, D=Draw, A=Away Win)


df_om_games.head()

df_om_games["OM_goals"] = np.where(df_om_games['HomeTeam'] == 'Marseille', df_om_games["FTHG"], df_om_games["FTAG"])
df_om_games["is_OM_win"] = np.where(df_om_games['HomeTeam'] == 'Marseille', df_om_games["FTR"] == "H", df_om_games["FTR"] == "A")

df_om_games.head()


# ADD GOALS RECEIVED
```
